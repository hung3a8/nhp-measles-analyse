```python
dataset = dataset.drop(columns = ['STT','ID','Name'])
dataset = dataset.drop(columns = ['SODIENTHOAI','Dianogis out of hospital','Fisrt of Department',
'Department of out hospital','Dianogis of test','Department of test'])
```

```python
dataset = dataset.rename(columns={"NGAYSINH.x":"dob"})
month_name = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
s = dataset['dob'].to_numpy()
a = 1
for i in range(len(s)):
    try:
        if '-' in s[i]:
            date = s[i].split(sep='-')
        elif '/' in s[i]:
            date = s[i].split(sep='/')
        for j in range(len(month_name)):
            if date[1] == month_name[j]:
                date[1] = j+1
        date = [int(x) for x in date]
        if date[2] < 2000:
            date[2] += 2000
        x = datetime.datetime(date[2],date[1],date[0])
        dataset.at[i,'dob'] = x
    except:
        a = 0
```

```python
comm = dataset['Community'].to_numpy(dtype='str')
for i in range(len(comm)):
    if comm[i] == "":
        continue
    s = ""
    for j in comm[i]:
        if (j>='a' and j<='z') or (j>='A' and j<='Z') or j ==' ' or (j>='0' and j<='9'):
            s += j
    dataset.at[i,'Community'] = s
dataset = dataset.rename(columns={"Community":"commune"})
```

```python
dataset = dataset.rename(columns={"date of admission":"admission_date"})
month_name = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
s = dataset['admission_date'].to_numpy()
a = 1
for i in range(len(s)):
    try:
        date = s[i].split(sep='-')
        for j in range(len(month_name)):
            if date[1] == month_name[j]:
                date[1] = j+1
        date = [int(x) for x in date]
        if date[2] < 2000:
            date[2] += 2000
        x = datetime.datetime(date[2],date[1],date[0])
        dataset.at[i,'admission_date'] = x
    except:
        a = 0
```
